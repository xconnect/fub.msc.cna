## Part 2 - Data Frames

In this exercise we will work with data frames. A data frame is similar to a table in a database. We prepared some code to import a csv. We advise to read and then execute the preimplemented code to understand. 

```r
# read the csv into a dataframe
path <- "data/msleep_ggplot2.csv"
df <- read.csv(path)

# show the head of the dataframe
head(df)
```

The data frame is now in the variable df. We can now analyze the data. To get a specific column or row we can use indices.
To get the first row we would execute `df[1,]`. To get the genus `df[,2]`. To get the genus of the first row: df[1,2].

**2.1 Get the name of all animals**

To get the rows of all animals who have less than 4 sleep_total: `df[df$sleep_total < 4,]`

### Masking

Lets understand how masking works. We use our previous vector 'n'.

**2.4 Assign a vector with the values [`TRUE, FALSE, TRUE, TRUE`] to the variable 'mask'. Now execute `n[mask]`**

```r
mask <- c(TRUE,FALSE,TRUE,TRUE);
n[mask];
```

We see that the second value from 'n' is ommitted coincidentally is only the second value of 'mask' FALSE.

Masking allows to take and filter multiple values of any dataframe or vector by supplying a boolean vector with TRUE values at the wanted indices. We can use this behaviour by obtaining this boolean vector with any boolean expression.

**2.5 Execute `n > 2`**

```r
n>2
```

The TRUE values are at the same indice where our condition holds TRUE (n>2).

By combining both we can obtain all values in n where n > 2.

**2.6 Execute n[n > 2]**

```r
n[n>2]
```
