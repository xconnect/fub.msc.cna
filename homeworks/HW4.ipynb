```r
G <- read.graph("data/sw.gml","gml")
```

```r
V(G)$color[V(G)$affiliation=="republic"] = 'lightblue'
V(G)$shape[V(G)$affiliation=="republic"] = 'circle'
V(G)$color[V(G)$affiliation=="empire"] = 'black'
V(G)$shape[V(G)$affiliation=="empire"] = 'square'

G <- delete.vertices(G, V(G)[V(G)$affiliation=="mixed"])
G <- delete.vertices(G, V(G)[V(G)$affiliation=="hutt"])
G <- delete.vertices(G, V(G)[V(G)$affiliation=="neutral"])

G <- delete.vertices(G, V(G)[degree(G) < 22])

plot.igraph(G, main = "Top 5 Star Wars heros ever, ever, ever...",
               vertex.size=V(G)$value/2, edge.color="red", vertex.label.color="blue",
               edge.width=E(G)$value/2, layout=layout.circle)
```

```r
G <- read.graph("data/sw.gml","gml")

V(G)$color[V(G)$affiliation=="hutt"] = 'yellow'
V(G)$shape[V(G)$affiliation=="hutt"] = 'circle'
V(G)$color[V(G)$affiliation!="hutt"] = 'black'
V(G)$shape[V(G)$affiliation!="hutt"] = 'square'
V(G)$size[V(G)$affiliation=="hutt"] = 10
V(G)$size[V(G)$affiliation!="hutt"] = 3
E(G)$width[E(G)$from=="hutt"]=E(G)$value/2

G <- delete.vertices(G, V(G)[V(G)$affiliation=="mixed"])
G <- delete.vertices(G, V(G)[V(G)$affiliation=="neutral"])
G <- delete.vertices(G, V(G)[V(G)$affiliation=="republic"])
G <- delete.vertices(G, V(G)[V(G)$affiliation=="empire"])
G_hutt <- delete.vertices(G, V(G)[degree(G) < 1])

plot.igraph(G_hutt, main = "The Hutt Connection", edge.color="yellow", edge.width=E(G)$value, layout = layout.circle)
```

```r
G <- read.graph("data/sw.gml","gml")

V(G)$color = 'white'
V(G)$shape = 'circle'

G <- delete.vertices(G, V(G)[V(G)$value1==0])
G <- delete.vertices(G, V(G)[V(G)$value2==0])
G <- delete.vertices(G, V(G)[V(G)$value3==0])
G <- delete.vertices(G, V(G)[V(G)$value4==0])
G <- delete.vertices(G, V(G)[V(G)$value5==0])
G <- delete.vertices(G, V(G)[V(G)$value6==0])

plot.igraph(G, main = "Evergreens", edge.color="blue", edge.width=E(G)$value, layout = layout.circle)
```
